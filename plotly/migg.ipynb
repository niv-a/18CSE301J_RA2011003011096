```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/niv-a/18CSE301J_RA2011003011096/main/mig.csv')
                 
fig = px.scatter(df, x="Total", y="Year", color="continent", marginal_y="violin",
           marginal_x="box", trendline="ols", template="simple_white", width=1000, height=600)
fig.update_traces(marker=dict(size=16,
                              line=dict(width=1,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

fig.update_layout(
    paper_bgcolor="LightSteelBlue",
)

fig.show()
```
